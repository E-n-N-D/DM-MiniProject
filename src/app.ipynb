```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)

songs = pd.read_csv('spotify_songs.csv')
songs.head()
songs.describe()

plt.figure(figsize=(12,6))
plt.hist(songs['duration_ms']/1000,bins=50,edgecolor='k')
plt.title('Songs Duration Distribution')
plt.xlabel('Duration (secs)')
plt.show()

songs['duration_ms'].mean()/1000




genre_tracks = songs.groupby('playlist_genre')['track_id'].nunique()

plt.pie(genre_tracks, labels=genre_tracks.index, autopct='%1.1f%%')
plt.title('Tracks per Genre')
plt.show()

artist_tracks = songs.groupby('track_artist')['track_id'].nunique().sort_values(ascending = False).head().sort_values()

plt.figure(figsize=(8,4))
artist_tracks.plot(kind='barh')
plt.title('Top 5 Artist by Track Number')
plt.xlabel('Tracks')
plt.ylabel('Artists')
plt.show()

playlist_track = songs.groupby('track_name')['playlist_id'].nunique().sort_values(ascending = False).head().sort_values()

plt.figure(figsize=(8,4))
playlist_track.plot(kind='barh')
plt.title('Top 5 Tracks in the Most Playlists')
plt.xlabel('Playlist')
plt.ylabel('Track Name')
plt.show()

artist_playlist = songs.groupby('track_artist')['playlist_id'].nunique().sort_values(ascending = False).head().sort_values()

plt.figure(figsize=(8,4))
artist_playlist.plot(kind='barh')
plt.title('Top 5 Artist by Playlist Number')
plt.xlabel('Playlists')
plt.ylabel('Artists')
plt.show()
```
